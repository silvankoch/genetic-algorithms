```python
import numpy as np
import matplotlib.pyplot as plt
from genetic_algorithm_base import GeneticAlgorithm
```

# Define population rules

```python
# In this example, an individual is a numpy array of 10 integers between 0 and 100
# The population is a list of 10 individuals
initial_population = [np.random.randint(0, 100, 10) for _ in range(10)]

# Define the fitness, crossover and mutation behavior
def fitness(individual):
    """Fitness of the individual"""
    return -sum(individual)**2

def crossover(parent1, parent2):
    """Cross over two parents to create a child"""
    crossover_point = np.random.randint(0, len(parent1))
    return np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))

def mutate(individual):
    """Mutate an individual by changing one of its genes"""
    mutation_point = np.random.randint(0, len(individual))
    individual[mutation_point] = np.random.randint(0, 100)
    return individual
```

# Evolving the population

```python
GA = GeneticAlgorithm(initial_population, fitness, crossover, mutate)

final_population, final_fitness, fitness_evolution = GA.evolve(number_of_generations=100, number_of_parents=5, number_of_offspring=10)

plt.plot(fitness_evolution)
plt.xlabel('Generation')
plt.ylabel('Fitness')
```
